#WHAT IS MACHINE LEARNING IN 100WORDS

###Machine learning is a sub-domain of computer science focused on algorithms that enable a computer to learn from data without being explicitly programmed to do so

Data can be numbers (like age or weight) or categories (like male/female)

##There are three main types:


*   supervised learning (learning from labeled examples)
*   unsupervised learning (finding hidden patterns in unlabeled data)
*   reinforcement learning (learning through rewards and penalties).


##Common models include
* linear regression,
* decision trees, and
* neural networks.

### To ensure accuracy, data is split into training and testing sets, and performance is measured using specific evaluation metrics and practical software tools.
