```python
import numpy as np
import pandas as pd

df = pd.read_csv("data.csv")

df.head()
```

# üîç Key Insights from the Dataset

### 1. Routine Consistency

Approximately 74% of the workouts lasted exactly 60 minutes.

This indicates a highly disciplined and structured exercise habit. Rather than random or sporadic activity, the data reflects a deliberate commitment to standardized training sessions.

Consistency is often a key predictor of long-term fitness progress.

```python
df.groupby("Duration")["Calories"].mean()
```

### 2. Duration Drives Calories

A clear relationship exists between workout duration and calories burned.

- 30-minute sessions average ~195 kcal
- 60-minute sessions average ~317 kcal

This reinforces a fundamental physiological principle: energy expenditure increases with time under activity.

However, duration alone does not fully explain calorie variation ‚Äî which leads to the next insight.

```python
df["Pulse"].mean()
```

### 3. Aerobic Zone Training

The overall average pulse across sessions is approximately 103 bpm.

This falls within the aerobic training zone for many adults, suggesting that the workouts were designed for cardiovascular endurance rather than maximum intensity or anaerobic bursts.

This reflects a focus on sustainable cardio conditioning rather than high-intensity intervals.

```python
df[df["Pulse"] > df["Maxpulse"]]
```

### 4. Data Quality Anomaly

One session (December 23) contains a physiologically impossible entry:
the recorded average pulse (130 bpm) exceeds the maximum pulse (101 bpm), This inconsistency highlights the importance of data validation before analysis. Even fitness tracking devices or exports can contain recording or entry errors.

Without identifying such anomalies, conclusions drawn from the dataset could be misleading.

```python
df[df["Duration"] == 60]["Calories"].min()

df[df["Duration"] == 60]["Calories"].max()
```

```python
df.groupby("Pulse")["Calories"].mean()
df[["Pulse","Calories"]].corr().corr()
```

### 5. pulse Intensity Influences Calorie Burn

Even among 60-minute sessions, calories burned range from 215 kcal to 479 kcal ‚Äî more than double.

This indicates that intensity, reflected by heart rate variation, significantly impacts energy expenditure.

Among sessions of equal duration, calorie burn increases with higher recorded pulse, suggesting a positive association between workout intensity and energy expenditure

## Overall Explanation

This dataset tells the story of a consistent fitness routine centered around  60-minute  sessions.

While the time invested is a major contributor to calorie expenditure, pulse intensity plays a critical role in differentiating workout outcomes.

The presence of a data anomaly also reinforces the importance of validating data before interpretation

Overall, the data reflects structured training behavior, moderate cardiovascular intensity, and meaningful variation driven by exertion levels.
